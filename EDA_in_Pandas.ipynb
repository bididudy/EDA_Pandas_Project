# Exploratory Data Analysis in pandas

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("/Users/bididudy/Downloads/Python/world_population.csv")
df
```

```python
# in this dataset float format has 1 digit after decimal point. We need to change to 2 digit after decimal points to look nice
pd.set_option('display.float_format', lambda x: '%.2f' %x)
df
```

```python
# we check the details of dataset
df.info()
```

```python
# there are no null values.
#int, object and float dataset are datatypes in dataset
#total 17 columns and 234 rows
```

```python
# to check count, mean, medium, mode, max, min 
df.describe()
```

```python
# to check all the null values numbers in the dataset
df.isnull().sum()
```

```python
# there are missing values in all population tables, area, density and growth tales.
```

```python
# to check unique values in dataset
df.nunique()
```

```python
# we can sort dataset by any column. here we are sorting by population 2022 and in decending order
df.sort_values(by='2022 Population', ascending=False).head()
```

```python
# we can check correlation in the dataset
df.corr()
```

```python
# with heatmap its easy to visualize correlation in dataset. darker it's getlower the correlation
sns.heatmap(df.corr(), annot=True) #annot shows the numbers on heatmap

plt.rcParams['figure.figsize'] = (20,7) #without setting figsize it wa looking smaller. we need to set figsize of the heatmap

plt.show() #to show the chart
```

```python
# populations are highly correlated with each other
#rank column has nothing to do with other columns
# world population percentage is highly correlated with all te population colmns
#area column is somehow correlated.
# density column is also correlated with populations tables
```

```python
# i am going to group by on continet and check mean of every column based on each continent.
df2=df.groupby('Continent')['1970 Population',
       '1980 Population', '1990 Population', '2000 Population',
       '2010 Population', '2015 Population', '2015 Population',
       '2022 Population'].mean().sort_values(by='2022 Population', ascending=False)
df2
```
